```python
import numpy as np
import pandas as pd
from time import time

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('IDEAs-TenantInfo-and-usage_shuffled.csv'#, nrows=200
                )
```

```python
df.head()
```

```python
cols_ar = df.columns.tolist()[-36:]
```

```python
cols_au = df.columns.tolist()[-78:-36]
drop = ['AU_oatp_04','AU_odsp_04','AU_oatp_05','AU_odsp_05','AU_oatp','AU_odsp']
cols_au = [x for x in cols_au if x not in drop]
```

```python
cols = list(zip(cols_ar, cols_au))
```

```python
print(cols[0][1])
```

```python
t1 = time()
for c in cols[:]:
    df[c[0]] = np.where(df[c[0]] != np.inf, df[c[0]], df[c[1]]/df['PaidCount'])
    print('Already complete coloumn {}'.format(c[0]))
t2 = time()
print(t2 - t1)
```

```python
df.replace(np.nan, 0, inplace=True)
```
