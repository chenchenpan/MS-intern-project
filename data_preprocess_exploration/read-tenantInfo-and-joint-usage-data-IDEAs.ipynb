```python
ar_list = list(zip(cols_ar, cols_au, cols_ps_multi))
```

```python
print(ar_list)
```

```python
def compute_ar(df, ar_list):
    for i in ar_list:
        df[i[0]] = df[i[1]] / df[i[2]]
    return df
```

```python
df_ar = compute_ar(df, ar_list)
```

```python
df_ar = shuffle(df_ar, random_state=0)
```
