```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv('TenantInfo-and-usage_shuffled_inf.csv')
```

```python
def clip_wl_ar_cols(df):    
    cols_name = pd.Series(data=df.columns)
    ar_04_beg_col_index = cols_name[cols_name == 'AR_exchange_04'].index[0]
    ar_06_beg_col_index = cols_name[cols_name == 'AR_exchange_06'].index[0]
    ar_06_end_col_index = cols_name[cols_name == 'AR_officeclient_06'].index[0]

    wl_AR_cols = cols_name[ar_04_beg_col_index:ar_06_end_col_index+1].tolist()

    output_cols = cols_name[ar_06_beg_col_index:ar_06_end_col_index+1].tolist()

    wl_arr = df.loc[:, wl_AR_cols].to_numpy()
    wl_clip = np.clip(wl_arr, 0.0, 1.0)

    df.loc[:, wl_AR_cols] = wl_clip
    
    return df
```

```python
df = clip_wl_ar_cols(df)
df.to_csv('TenantInfo-and-usage_shuffled_inf_clip.csv', index=None)
```

```python
def split_mature_and_young(df, age=360):
    df_mature = df.loc[df['Age'] >= age]
    df_young = df.loc[df['Age'] < age]
    
    df_mature = df_mature.drop('Age', axis=1)
    df_young = df_young.drop('Age', axis=1)
    
    print('full dataset shape is {}'.format(df.shape))
    print('df_young shape is {}'.format(df_young.shape))
    return df_mature, df_young
```

```python
df_clip = pd.read_csv('TenantInfo-and-usage_shuffled_inf_clip.csv')
```

```python
df_c_mature, df_c_young = split_mature_and_young(df_clip)    
```

```python
df_c_young = df_c_young.sample(frac=1.0)
df_c_young.to_csv('TenantInfo-and-usage_shuffled_inf_clip_young.csv', index=None)
```

```python
df_c_mature = df_c_mature.sample(frac=1.0)
df_c_mature.head()
```

```python
df_c_mature.to_csv('TenantInfo-and-usage_shuffled_inf_clip_mature.csv', index=None)
```
